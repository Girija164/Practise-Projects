```python
for item in search_items:
    # Find the search bar
    srchBar = driver.find_element_by_xpath("//div[@class='_3OO5Xc']/input")
    srchBar.clear()
    srchBar.send_keys(item)
    
     # Clicking on the search button
    srcbtn= driver.find_element_by_xpath("//button[@class='L0Z3Pu']").click()
    time.sleep(3)
    
    url_tag= driver.find_elements_by_xpath("//a[@class='_1fQZEK']")
    for url in url_tag:
        urls.append(url.get_attribute('href'))
    
```

# Amazon Website

```python
driver= webdriver.Chrome(r'C:\Users\user\Downloads\chromedriver_win32 (3)\chromedriver.exe')
```

```python
url = "https://www.amazon.in/"
driver.get(url)
time.sleep(3)
```

```python
driver.maximize_window()
time.sleep(2)
```

```python
title = []
review_text = []
ratings = []
url1=[]
```

```python
search_items = ['laptops','phones','headphones','smart watches', 'Professional Cameras', 'Printers', 
              'Monitors', 'Home theater', 'Router']
```

```python
for item in search_items:
    #To access search bar
    searchbar = driver.find_element_by_id("twotabsearchtextbox")
    searchbar.clear()
    searchbar.send_keys(item)
    
     #clicking on search button
    searchbtn=driver.find_element_by_id("nav-search-submit-button").click() 
    time.sleep(1)
    
    url1_tag= driver.find_elements_by_xpath("//h2[@class='a-size-mini a-spacing-none a-color-base s-line-clamp-2']/a")
    for url in url1_tag:
        url1.append(url.get_attribute('href'))
```

```python
len(url1)
```

```python
for i in url1[180:227]:
    driver.get(i)
    time.sleep(2)
    
    #Clicking the rating
    try:
        driver.find_element_by_xpath("//a[1][@id='acrCustomerReviewLink']").click()
        #Clicking to see all reviews
        try:
            driver.find_element_by_xpath("//a[@class='a-link-emphasis a-text-bold']").click()
        except NoSuchElementException:
            pass
    except NoSuchElementException:
        print("No rating")
        pass
    
    
    
    #Scrapping the details
    Start_page=1
    End_page=10
    for page in range(Start_page,End_page+1):
        try:
            Reviews=driver.find_elements_by_xpath("//div[@class='a-row a-spacing-small review-data']/span/span")
            for r in Reviews:
                review_text.append(r.text.replace('\n',''))
        except NoSuchElementException:
            review_text.append("Not Available")
        try:
            Rating=driver.find_elements_by_xpath("//div[@class='a-section celwidget']/div[2]/a[1]")
            for i in Rating:
                rating=i.get_attribute('title')
                ratings.append(rating[:1])
        except NoSuchElementException:
            ratings.append("Not available") 
        
        try:
            title_tag=driver.find_elements_by_xpath("//div[@class='a-section celwidget']/div[2]/a[2]")
            for titles in title_tag:
                title.append(titles.text)
        except NoSuchElementException:
            title.append("Not available")
        
           
        #Looping for next page automatically
        try:
            next_page=driver.find_element_by_xpath("//div[@id='cm_cr-pagination_bar']/ul/li[2]/a")
            if next_page.text=='Next Page':
                next_page.click()
                time.sleep(2)
        except NoSuchElementException:
            pass
```
